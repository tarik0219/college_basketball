```python
er.drop(['away'], axis = 1, inplace = True)
df = pd.merge(er, kp, how='left', left_on=['home','date'], right_on=['home','date'])
df
```

```python
for index, row in df.iterrows():
    line = row['draft_kings']
    score = row['score']
    if score > line:
        win = 1
    elif score < line:
        win = 0
    else:
        win = 2
    
    
    df.loc[index,'cover'] = win
df
```
