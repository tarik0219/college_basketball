```python
import pandas as pd
```

```python
colnames=['YEAR', 'TEAM', 'OFF', 'DEF','TEMPO','LUCK'] 
kp = pd.read_csv('kenpom.csv', names=colnames, header=None)
kp.head()
```

```python
for i in range(len(kp)):
    kp.loc[i,"TEAM"] = kp.loc[i,"TEAM"].lower()
kp.head()
```

```python
punctuations = '.'

for i in range(len(kp)):
    team = kp.loc[i,"TEAM"]
    no_punct = ""
    for char in team:
       if char not in punctuations:
           no_punct = no_punct + char
    
    kp.loc[i,"TEAM"] = no_punct
```

```python
colnames=['YEAR', 'SCORE_1','TEAM_1','TEAM_2','SCORE_2'] 
df = pd.read_csv('Big_Dance_CSV.csv', names=colnames, header=None)
df.head()
```

```python
import numpy as np
df['OFF_1' ] = np.nan
df['DEF_1' ] = np.nan
df['TEMPO_1' ] = np.nan
df['LUCK_1' ] = np.nan
df['OFF_2' ] = np.nan
df['DEF_2' ] = np.nan
df['TEMPO_2' ] = np.nan
df['LUCK_2' ] = np.nan

```

```python
df.head()
```

```python
for i in range(len(df)):
    df.loc[i,"TEAM_1"] = df.loc[i,"TEAM_1"].lower()

punctuations = '.'

for i in range(len(df)):
    team = df.loc[i,"TEAM_1"]
    no_punct = ""
    for char in team:
       if char not in punctuations:
           no_punct = no_punct + char
    
    df.loc[i,"TEAM_1"] = no_punct
df.head()
```

```python
for i in range(len(df)):
    df.loc[i,"TEAM_2"] = df.loc[i,"TEAM_2"].lower()

punctuations = '.'

for i in range(len(df)):
    team = df.loc[i,"TEAM_2"]
    no_punct = ""
    for char in team:
       if char not in punctuations:
           no_punct = no_punct + char
    
    df.loc[i,"TEAM_2"] = no_punct
df.head()
```

```python
for i in range(len(df)):
    team = df.loc[i,"TEAM_1"]
    year = df.loc[i,"YEAR"]
    try:
        test = kp.loc[(kp['TEAM'] == team ) & (kp['YEAR'] == year)]
        for index, rows in test.iterrows(): 
            my_list =[rows.OFF, rows.DEF, rows.TEMPO,rows.LUCK]
        df.loc[i,"OFF_1"] = my_list[0]
        df.loc[i,"DEF_1"] = my_list[1]
        df.loc[i,"TEMPO_1"] = my_list[2]
        df.loc[i,"LUCK_1"] = my_list[3]
    except:
        continue
df.head()
        
```

```python
for i in range(len(df)):
    team = df.loc[i,"TEAM_2"]
    year = df.loc[i,"YEAR"]
    try:
        test = kp.loc[(kp['TEAM'] == team ) & (kp['YEAR'] == year)]
        for index, rows in test.iterrows(): 
            my_list =[rows.OFF, rows.DEF, rows.TEMPO,rows.LUCK]
        df.loc[i,"OFF_2"] = my_list[0]
        df.loc[i,"DEF_2"] = my_list[1]
        df.loc[i,"TEMPO_2"] = my_list[2]
        df.loc[i,"LUCK_2"] = my_list[3]
    except:
        continue
df.head()
```
