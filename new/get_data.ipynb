```python
score = []
for i,x in enumerate(predict):
    team = home[i]
    list_1 = x.split()
    team_compare = ''
    num = len(list_1) - 2
    
    for i in range(0,num):
        team_compare = team_compare + ' '+ list_1[i]
    team_compare = team_compare[1:]
    if team_compare == team:
        list_2 = list_1[-2].split('-')
        score.append(int(list_2[0]) - int(list_2[1]))
    else:
        list_2 = list_1[-2].split('-')
        score.append(int(list_2[1]) - int(list_2[0]))
```
