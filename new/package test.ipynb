```python
from datetime import datetime
from datetime import timedelta
import os
path = os.getcwd()
import pandas as pd
def get_yesterday():
    today = datetime.today()
    yesterday = today - timedelta(days = 1)
    year = yesterday.year
    month = yesterday.month
    day = yesterday.day
    return year,month,day
```

```python
import kenpompy.FanMatch as fm
```

```python
fm
```

```python
year,month,day = get_yesterday()
test = fm.FanMatch(browser, date = str(year)+'-'+str("{0:0=2d}".format(month))+'-'+str("{0:0=2d}".format(day)))
```

```python
df = test.fm_df
df
```

```python
wt = df['Winner'].tolist()
lt = df['Loser'].tolist()
ws = df['WinnerScore'].tolist()
ls = df['LoserScore'].tolist()
```

```python
wt
```

```python
year,month,day = get_yesterday()
name = path+"\\games\\"+str(year)+"_"+str(month)+"_"+str(day)+"_games.csv"
scores = pd.read_csv(name)
```

```python
scores['home_score'] = -99
scores['away_score'] = -99
scores
```

```python
for index, row in scores.iterrows():
    home = row['home']
    away = row['away']
    for i,x in enumerate(wt):
        if home == x:
            scores.loc[index,'home_score'] = int(ws[i])
        if away == x:
            scores.loc[index,'away_score'] = int(ws[i])
        if home == lt[i]:
            scores.loc[index,'home_score'] = int(ls[i])
        if away == lt[i]:
            scores.loc[index,'away_score'] = int(ls[i])
```

```python
scores = scores.drop(columns=['predict_kp'])
```

```python
scores = scores[scores['home_score']!=-99]
scores['score'] = scores['home_score'] - scores['away_score']
scores['date'] = str(year)+'-'+str("{0:0=2d}".format(month))+'-'+str("{0:0=2d}".format(day))
scores
```

```python
scores.to_csv('df.csv', index=False)
```
